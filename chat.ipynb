### Librerias

```python
import PyPDF2 # Extraccion de texto

# Prosesamiento de texto
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

## 1. Extracción de texto:

```python
def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

# Ruta del archivo PDF
pdf_file_path = './pdf/constitucion_bolivia.pdf'
pdf_text = extract_text_from_pdf(pdf_file_path)
```

## 2.- Prosesamiento de texto

```python
def preprocess_text(text):
    # Tokenización
    tokens = word_tokenize(text)
    # Eliminación de stopwords y puntuación
    stop_words = set(stopwords.words('english') + list(string.punctuation))
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

processed_text = preprocess_text(pdf_text)
```

## 3.- chatbot

```python
# Dividir el texto en artículos
articles = pdf_text.split("Artículo")

# Procesar cada artículo
for article in articles:
    # Ejemplo: buscar palabras clave en el artículo
    if 'vida' in article:
        # Si la palabra clave está en el artículo, imprimir el artículo
        print("Artículo encontrado:", article)

def search_article_by_keyword(keyword):
    # Buscar un artículo específico por palabra clave
    found_articles = [article for article in articles if keyword in article]
    return found_articles

# Ejemplo de búsqueda y respuesta a preguntas
keyword = 'hola que tal'
matching_articles = search_article_by_keyword(keyword)

if matching_articles:
    for article in matching_articles:
        print("Artículo encontrado:", article)
else:
    print("No se encontraron artículos relacionados.")
```
